```python
# Contiene información técnica y de mercado de automóviles. El objetivo es predecir 
# el precio del automóvil en función de sus atributos

# Realizar las diferentes gráficas que permitan visualizar los resultados del modelo.

# Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
data = pd.read_csv('Dataset_Regresion_Lineal/Car details v3.csv')

# Limpieza de datos (si es necesario)
data = data.dropna()  # Eliminar filas con valores nulos

# 1. Histograma del precio de venta
plt.figure(figsize=(12, 6))
sns.histplot(data['selling_price'], bins=30, kde=True)
plt.title('Distribución del Precio de Venta')
plt.xlabel('Precio de Venta')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# 2. Gráfica de dispersión: Potencia del motor vs Precio
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x='power', y='selling_price')
plt.title('Potencia del Motor vs Precio de Venta')
plt.xlabel('Potencia del Motor')
plt.ylabel('Precio de Venta')
plt.grid(True)
plt.show()

# 3. Boxplot de precios por marca
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='brand', y='selling_price')
plt.title('Distribución de Precios por Marca')
plt.xlabel('Marca')
plt.ylabel('Precio de Venta')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x
plt.grid(True)
plt.show()

# 4. Mapa de calor de correlación
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Mapa de Calor de Correlaciones')
plt.show()


```
